```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix, accuracy_score, mean_absolute_error, mean_squared_error, r2_score
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import pickle

# Load the data
df = pd.read_csv('/content/T1.csv')
print(df.describe())

# Data preprocessing
df1 = df.copy()
time_format = "%d %m %Y %H:%M"
df1["Date/Time"] = df1["Date/Time"].astype(str).apply(lambda x: datetime.strptime(x, time_format))
df1.set_index('Date/Time', inplace=True)
```

```python
# Plot distribution of dataset variables
plt.figure(figsize=(15,8))
plt.suptitle('Distribution of Dataset Variables', fontsize=15)
for i in range(1, len(df.columns)):
    plt.subplot(2, 2, i)
    fig = df[df.columns[i]].hist(bins=20)
    fig.set_xlabel(df.columns[i], fontsize=12)
    if (i % 2 != 0):
        fig.set_ylabel('Number of Values', fontsize=12)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()

# Scatter plots for Wind Direction
n = 0
plt.figure(figsize=(20,40))
for i in range(0, 360, 10):
    condition = (df1['Wind Direction (째)'] > i) & (df1['Wind Direction (째)'] < i + 10)
    n += 1
    fig = plt.subplot(12, 6, n)
    plt.scatter(df1.loc[condition, 'Wind Speed (m/s)'], df1.loc[condition, 'LV ActivePower (kW)'])
    plt.scatter(df1.loc[condition, 'Wind Speed (m/s)'], df1.loc[condition, 'Theoretical_Power_Curve (KWh)'])
    fig.set_title(f'{i}<Wind Direction (째)<{(i+10)}')
    plt.subplots_adjust(hspace=0.4)
plt.show()

# Defining Power deficit values in percentage
df2 = df1.copy()
power_deficit_percentage = 100 * (df2['LV ActivePower (kW)'] - df2['Theoretical_Power_Curve (KWh)']) / df2['Theoretical_Power_Curve (KWh)']
df2['power_deficit_percentage'] = power_deficit_percentage
df2.loc[df2['Theoretical_Power_Curve (KWh)'] == 0, 'power_deficit_percentage'] = 0

# Defining the status label of the turbines with two numeric values: 1 for "ok"; 0 for "not ok"
threshold = 10
condition = (df2["power_deficit_percentage"] < threshold) & (df2["power_deficit_percentage"] > -threshold)
df2.loc[condition, "turbine_status"] = 1
condition = (df2["power_deficit_percentage"] > threshold) | (df2["power_deficit_percentage"] < -threshold)
df2.loc[condition, "turbine_status"] = 0
```

```python
# Prepare data for modeling
x = df2.drop(columns=["turbine_status"])
y = df2['turbine_status']
scaler = preprocessing.StandardScaler().fit(x)
x = scaler.transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Define function to evaluate models
def evaluate_model(model, x_train, y_train, x_test, y_test):
    # Cross-validation
    cv_scores = cross_val_score(model, x_train, y_train, cv=5)
    print(f'Cross-Validation Scores: {cv_scores}')
    print(f'Mean CV Score: {np.mean(cv_scores):.2f}')

    # Train and predict
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    # Evaluation
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy: {accuracy * 100:.2f}%')
    print("Mean Absolute Error:    ", mean_absolute_error(y_test, y_pred))
    print("Mean Squared Error:     ", mean_squared_error(y_test, y_pred))
    print("R2 Score:               ", r2_score(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    cm_matrix = pd.DataFrame(data=cm, columns=['Actual Not Ok', 'Actual Ok'], index=['Predict Not Ok', 'Predict Ok'])
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='Blues', linewidths=.5)
    plt.title(f'Confusion Matrix for {type(model).__name__}')
    plt.show()

# Logistic Regression
log_reg = LogisticRegression(max_iter=1000)
evaluate_model(log_reg, x_train, y_train, x_test, y_test)

# MLP Classifier (Deep Learning)
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, solver='adam', random_state=1)
evaluate_model(mlp, x_train, y_train, x_test, y_test)

# Support Vector Machine
svc = SVC(kernel='linear')
evaluate_model(svc, x_train, y_train, x_test, y_test)

# Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=1)
evaluate_model(rf, x_train, y_train, x_test, y_test)

# Choose the best model based on performance (Assume RandomForest is the best for demonstration)
best_model = RandomForestClassifier(n_estimators=100, random_state=1)
best_model.fit(x_train, y_train)
y_pred = best_model.predict(x_test)

# Final Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f'Final Model Accuracy: {accuracy * 100:.2f}%')
print("Mean Absolute Error:    ", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:     ", mean_squared_error(y_test, y_pred))
print("R2 Score:               ", r2_score(y_test, y_pred))

# Final Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Not Ok', 'Actual Ok'], index=['Predict Not Ok', 'Predict Ok'])
plt.figure(figsize=(8, 6))
sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='Blues', linewidths=.5)
plt.title('Final Confusion Matrix for Best Model (Random Forest)')
plt.show()

# Save the final model to a pickle file
with open('final_model.pkl', 'wb') as file:
    pickle.dump(best_model, file)
```
